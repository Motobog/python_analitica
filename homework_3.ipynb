### Задача 1
Скачать файл kc_house_data.csv - приложен на платформе</br>
Считать данные с помощью pandas</br>
Вывести на экран первые 5 строк

```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv')
df.head()
```

#### Задача 1.2
Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома</br>
Если реновации дома не было, то в новом признаке поставьте 0



```python
df['delta_renovated'] = 0
df.loc[df['yr_renovated']!=0,'delta_renovated'] = df['yr_renovated'] - df['yr_built']
df.head()
```

#### Задача 1.3
Создайте признаки года продажи, месяца продажи

```python
df['yr_sale'] = df['date'].apply(lambda x: int(str(x)[:4]))
df['month_sale'] = df['date'].apply(lambda x: int(str(x)[4:6]))
df.head()
```

#### Задача 1.4
Удалите признаки date, zipcode, lat, long


```python
df.drop(columns = ['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```

### Задача 3

#### Задача 3.1
Найдите среднюю стоимость домов в зависимости от количества спален</br>
Отсортируйте от меньшей стоимости к большей

```python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values(by = 'price')
```

#### Задача 3.2
Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['min','mean', 'max']})
```

#### Задача 3.3
